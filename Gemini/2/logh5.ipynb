```python
import tensorflow as tf
import numpy as np
import time
import os
import glob
from PIL import Image

# Cấu hình
MODEL_PATH = "best_model.h5"
TEST_DIR = "C:/DUT/Ki 2/Edge AI/dataset/test"
IMG_SIZE = (32, 32)

print("Đang tải mô hình H5...")
model = tf.keras.models.load_model(MODEL_PATH)

# Lấy 700 ảnh từ tập test
test_image_paths = glob.glob(os.path.join(TEST_DIR, "*.[jp][pn]*"))
test_image_paths.sort()
test_samples = test_image_paths[:700] 

print(f"Tổng số ảnh Test để test: {len(test_samples)}")
print("Bắt đầu chạy suy luận (Inference)...")

total_inference_time = 0
# Giả sử bạn có nhãn để test (nếu không có nhãn, Accuracy sẽ mặc định 100% hoặc bỏ qua)
correct = 0 

for img_path in test_samples:
    img = Image.open(img_path).convert('RGB').resize(IMG_SIZE)
    img_array = np.expand_dims(np.array(img) / 255.0, axis=0)
    
    start_time = time.time()
    res = model.predict(img_array, verbose=0)
    end_time = time.time()
    
    total_inference_time += (end_time - start_time)
    
    # Giả định Id file khớp với Label để test Accuracy (nếu có)
    # predicted = np.argmax(res)
    # if predicted == true_label: correct += 1

# Tính toán thông số
avg_time_ms = (total_inference_time / len(test_samples)) * 1000

print("="*40)
print("📊 BÁO CÁO ĐÁNH GIÁ MÔ HÌNH H5 (FLOAT32)")
print("="*40)
# Lưu ý: Accuracy này chỉ chính xác nếu bạn có nhãn thực tế để so sánh
print(f"🎯 Độ chính xác (Accuracy): [Cần nhãn để tính]%") 
print(f"⏱️ Tổng thời gian chạy:     {total_inference_time*1000:.4f} ms")
print(f"⚡ Tốc độ trung bình:       {avg_time_ms:.2f} ms / ảnh")
print("="*40)
```
