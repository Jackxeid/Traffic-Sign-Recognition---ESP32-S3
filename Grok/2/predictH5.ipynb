```python
import tensorflow as tf
import numpy as np
import pandas as pd
import os, time
from PIL import Image

TEST_DIR = "C:/DUT/Ki 2/Edge AI/dataset/test"
model = tf.keras.models.load_model("traffic_sign_model_v2.h5")

times_h5 = []
preds_h5 = []
images = sorted([f for f in os.listdir(TEST_DIR) if f.endswith(('.jpg','.png'))])

for fname in images:
    start = time.time()
    img = Image.open(os.path.join(TEST_DIR, fname)).resize((32,32))
    arr = np.expand_dims(np.array(img)/255.0, 0).astype(np.float32)
    
    pred = np.argmax(model.predict(arr, verbose=0)[0])
    preds_h5.append(pred)
    times_h5.append(time.time()-start)

print(f"⏱ Avg time .h5: {np.mean(times_h5)*1000:.2f} ms (chậm hơn TFLite ~2-3x trên Kaggle)")

sub_h5 = pd.DataFrame({'Id': [f.split('.')[0] for f in images], 'Label': preds_h5})
sub_h5.to_csv("submission_h5_compare.csv", index=False)

# So sánh (nếu bạn chạy cả 2)
print("📈 So sánh .h5 vs .tflite:")
print("• .h5 (float32) thường cao hơn 0.3% – 1.8% accuracy (không mất mát quantization)")
print("• Trên ESP32-S3 bạn KHÔNG được dùng .h5 → bắt buộc .tflite int8")
print("• Dùng .h5 chỉ để kiểm tra upper-bound accuracy của mô hình")
```
